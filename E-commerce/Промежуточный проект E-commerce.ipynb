# 5. Напиши функцию на python, позволяющую строить когортный анализ. В период с января по декабрь выяви когорту с самым высоким retention на 3-й месяц. (10 баллов)

Для начала берем исходные данные по клиентам и заказам и формируем из них новый датафрейм для задания. Затем добавляем колонку с датой заказа в формате `'Год-месяц'`.

```python
cogorts = orders.merge(customers, on='customer_id')[['customer_unique_id', 'order_purchase_timestamp']]
cogorts['order_month'] = cogorts.order_purchase_timestamp.dt.to_period('M')
cogorts
```

```python
cogorts = cogorts.merge(cogorts.groupby('customer_unique_id').order_month.min(), on='customer_unique_id', how='left')
cogorts.rename(columns={'order_month_x': 'order_month', 'order_month_y': 'first_order_month'}, inplace=True)
cogorts
```

Построим когорты:

```python
grouped_cogorts = cogorts.groupby(['first_order_month', 'order_month']).agg(
    {
        'customer_unique_id': pd.Series.nunique,
        'order_purchase_timestamp': pd.Series.count
    }).rename(columns={
    'customer_unique_id': 'total_customers',
    'order_purchase_timestamp': 'orders_count'
})
grouped_cogorts
```

```python
def CogortPeriod(df):
    df['cohort_period'] = np.arange(0, len(df))
    return df
grouped_cogorts = grouped_cogorts.groupby(level=0).apply(CogortPeriod)
grouped_cogorts
```

```python
grouped_cogorts.reset_index(inplace=True)
grouped_cogorts.set_index(['cohort_period', 'first_order_month'], inplace=True)
```

```python
grouped_cogorts['total_customers'].unstack(0)
```

```python
cohort_group_size = grouped_cogorts['total_customers'].groupby(level=1).first()
cohort_group_size
```

```python
retention_rate = grouped_cogorts['total_customers'].unstack(0).divide(cohort_group_size, axis=0)
retention_rate
```

Посмотрим на итоговый **retention** и обратим внимание на колонку под номером **3**. Она и будет показывать **retention** на третий месяц:

```python
cohort_analysis = (retention_rate
            .style
            .set_caption('User retention by cohort') 
            .background_gradient(cmap='icefire')
            .highlight_null('white') 
            .format("{:.2%}", na_rep=""))
cohort_analysis
```

Самый высокий **retention** на третий месяц получился для пользователей которые пришли к нам **2017-06**

# 6. Построй RFM-кластеры для пользователей. Выведи для каждого кластера средние значения метрик R, F, M (и опиши подробно, как были построены метрики R, F, M). (35 баллов)

Для начала соеденим необходимые таблицы и отберем нужные колонки:

```python
user_info = orders.merge(customers, on='customer_id')[['customer_unique_id', 'order_purchase_timestamp', 'order_id', 'customer_id']]
user_info
```

```python
full_info = user_info.merge(order_items, on='order_id', how='left')[['customer_unique_id', 'order_purchase_timestamp', 'order_id', 'customer_id', 'price']]
full_info
```

```python
date_max = max(full_info.order_purchase_timestamp) + datetime.timedelta(days=1)
date_max
```

Начнем считать все необходимые метрики!

`Recency` - дельта, которая показывает, когда была совершена последняя покупка. Чем этот показатель ниже, тем нам лучше.

`Frequency` - считаем число заказов на одного пользователя.

`Monetary` - считаем суммарную сумму, потраченную пользователем.

```python
metrics = full_info.groupby('customer_unique_id').agg({
    'order_purchase_timestamp': lambda x: (date_max - x.max()).days,
    'order_id': pd.Series.count,
    'price': pd.Series.sum
}).rename(columns={
    'price': 'Monetary',
    'order_id': 'Frequency',
    'order_purchase_timestamp': 'Recency',
})
metrics
```

Разобьем метрики на 5 квантилей и посчитаем RFM-кластеры для каждого пользователя

```python
quantiles = metrics.quantile(q=[0.20, 0.40, 0.60, 0.80])
quantiles = quantiles.to_dict()

def RScore(x,p,d):
     if x <= d[p][0.20]:
         return 1
     elif x <= d[p][0.40]:
         return 2
     elif x <= d[p][0.60]: 
         return 3
     elif x<=d[p][0.80]:
         return 4
     else:
         return 5
```

```python
metrics['r_quartile'] = metrics['Recency'].apply(RScore, args=('Recency',quantiles,))
metrics['f_quartile'] = metrics['Frequency'].apply(RScore, args=('Frequency',quantiles,))
metrics['m_quartile'] = metrics['Monetary'].apply(RScore, args=('Monetary',quantiles,))
metrics['RFMScore'] = metrics.r_quartile.map(str) + metrics.f_quartile.map(str) + metrics.m_quartile.map(str)
metrics
```

```python
metrics[['Recency', 'Frequency', 'Monetary', 'RFMScore']].groupby('RFMScore').mean()
```

```python
metrics[['Recency', 'Frequency', 'Monetary', 'RFMScore']].groupby('RFMScore').Recency.mean().plot(kind='bar')
```

```python
metrics[['Recency', 'Frequency', 'Monetary', 'RFMScore']].groupby('RFMScore').Frequency.mean().plot(kind='bar')
```

```python
metrics[['Recency', 'Frequency', 'Monetary', 'RFMScore']].groupby('RFMScore').Monetary.mean().plot(kind='bar')
```

Мы получили RFM-кластеры на основании которых в будующем можно будет работать с каждым кластером отдельно: предлагать различные бонусы, акции, скидки, специальные предложения и т.д.
